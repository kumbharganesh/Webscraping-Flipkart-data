```python
import pandas as pd
import requests
from bs4 import BeautifulSoup
```

```python
site=requests.get('https://www.flipkart.com/search?q=tv&as=on&as-show=on&otracker=AS_Query_TrendingAutoSuggest_8_0_na_na_na&otracker1=AS_Query_TrendingAutoSuggest_8_0_na_na_na&as-pos=8&as-type=TRENDING&suggestionId=tv&requestId=9c9fa553-b7e5-454b-a65b-bbb7a9c74a29')
print(site)
```

```python
print(site.content)
```

```python
data=BeautifulSoup(site.content,'html.parser')
print(data.prettify())
```

```python
productname=data.find_all('div',class_='_4rR01T')
print(productname)
```

```python
for i in data.find_all('div',class_="_4rR01T"):
    productname=i.text
    print(productname)
```

```python
Title=[]
for a in data.find_all('div',class_="_4rR01T"):
    productname=a.text
    Title.append(productname)
print(Title)
```

```python
Ratings=[]
for i in data.find_all('div',class_='_3LWZlK')[0:24]:
    ratingname=i.text
    Ratings.append(ratingname)
print(Ratings)
```

```python
Price=[]
for b in data.find_all('div',class_='_30jeq3 _1_WHN1'):
    price1=b.text
    Price.append(price1)
print(Price)
```

```python
print(len(Title))
print(len(Ratings))
print(len(Price))
```

```python
myflipkartdata=pd.DataFrame({'Product Name':Title,'Price':Price,'Ratings':Ratings})
myflipkartdata
```

```python
myflipkartdata.to_csv('wescraping_Flipkart_.csv',index=False)
```
